```python
from openai import OpenAI 
```

```python
from autogen import ConversableAgent
```

```python
llm_config = {
    "model": "gpt-4o-mini", # From the list of models provided by OpenAI https://platform.openai.com/docs/models/continuous-model-upgrades
    "api_key": "YoUR_OPENAI_API_KEY"
    }
```

```python
socrates = ConversableAgent(
    name="socrates",
    llm_config=llm_config,
    system_message=(
        "You are Socrates, the ancient Greek philosopher. "
        "Your role is to challenge ideas by asking deep, thought-provoking questions. "
        "Use the Socratic method. When you are ready to end the conversation, say 'Thank you for having me'."
    ),
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "Thank you for having me" in msg["content"],
)

tesla = ConversableAgent( 
    name="tesla", 
    llm_config=llm_config, 
    system_message=(
        "You are Nikola Tesla, the visionary inventor and electrical engineer. "
        "You speak with passion about science, innovation, and the limitless potential of electricity. "
        "You often share groundbreaking ideas with confidence, even if they seem unconventional. "
        "When you are ready to end the conversation, say 'The future will understand me'."
    ), 
    human_input_mode="NEVER",
    is_termination_msg=lambda msg: "The future will understand me" in msg["content"],
)

    
```

```python
chat_result = tesla.initiate_chat(
    recipient=socrates,
    message=(
        "I am Nikola Tesla, and I am honored to engage in this discussion on the nature of knowledge and human progress. "
        "Socrates, I wonder—do you believe that relentless questioning alone can lead to true scientific breakthroughs, "
        "or must there also be a spark of intuition and vision?"
    ),
)
```
